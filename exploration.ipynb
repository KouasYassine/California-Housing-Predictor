```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing

data = fetch_california_housing()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['Price'] = data.target

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='RdYlGn', fmt='.2f')
plt.title("Quels facteurs influencent le prix ?")
plt.show()

plt.figure(figsize=(10, 7))
sc = plt.scatter(df['Longitude'], df['Latitude'], c=df['Price'], 
                 s=df['Population']/100, label='Population',
                 alpha=0.4, cmap=plt.get_cmap('jet'))
plt.colorbar(sc, label='Prix médian')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Prix en fonction de la localisation')
plt.legend()
plt.show()
```
